```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

```python
df = pd.read_csv('cluster_data.csv')
```

```python
cols = ['price_per_sqm', 'building_age', 'building_size']
df_prepared = df.copy()

for col in ['price_per_sqm', 'building_age', 'building_size']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]

features = ['price_per_sqm', 'building_age', 'building_size']
X = df_prepared[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(20, 12))

for i, col in enumerate(features, 1):
    # Boxplot
    plt.subplot(2, len(features), i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')

    # Histogram + KDE
    plt.subplot(2, len(features), i + len(features))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Histogram & KDE of {col}')

plt.tight_layout()
plt.show()
```
